Q2 - In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a data frame and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“.

```python
from time import sleep 
page=0

product=[]
rating=[]
no_rating=[]
price=[]
ex=[]
delivery=[]
availability=[]
details=[]
product_url=[]

while page<3:
    sleep(5)
    
    #scraping product name 
    p=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']")
    for i in p:
        product.append(i.text)
        
    
    #scraping product url 
    url=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a")
    for i in url:
        product_url.append(i.get_attribute("href"))
        
   
            
    
            
    page=page+1 #page increment 
    
    #to go to the next page
    n=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    n.click()
```

```python
for j in product_url:   
    driver.get(j)
    
    #scraping price
    try:
        pr=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price.append(pr.text)
    except:
        price.append("--")
     
    #scraping rating
    try:
        r=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(r.text)
    except:
        rating.append("--")
        
    
    #scraping no. of ratings    
    try:
        nr=driver.find_element_by_xpath("//div[@class='a-row a-spacing-medium averageStarRatingNumerical']")
        no_rating.append(nr.text)
    except:
        no_rating.append("--")
        
    
    #scraping avaliablity of product
    try:
        a=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        availability.append(a.text)
    except:
        availability.append("--")
        
    
    #scraping excepted delivery date
    try:
        d=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']//b")
        delivery.append(d.text)
    except:
        delivery.append("--")
        
        
    #scraping return/exchange
    try:
        e=driver.find_elements_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")[1]
        ex.append(e.text)
    except:
        ex.append("--")
        
        
    #scraping details of the product    
    try:
        d=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        details.append(d.text)
    except:
        details.append("--")
    
    
```

```python
import pandas as pd 

#making a dataframe 
data=list(zip(product,
rating,
no_rating,
price,
ex,
delivery,
availability,
details,
product_url,))

amazon_books=pd.DataFrame(data=data,columns=["Product_name","Ratings","No. of Rating","Price","Exchange/return","Delivery","Availability","Details","Product_url"])
amazon_books
```

Q3 - Write a python program to access the search bar and search button on images.google.com and 
scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
driver=webdriver.Chrome(r"C:\Users\Hp\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
def search(x):
    driver.get("https://images.google.com/")
    s_bar=driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    s_bar.send_keys(x)    # searching for keywords
    s_button=driver.find_element_by_xpath("//span[@class='z1asCe MZy1Rb']")
    s_button.click()
    
    #get scrolling height
    l_ht= driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #scroll to the bottom of the page
        sleep(5)
        n_ht = driver.execute_script("return document.body.scrollHeight")
    #comparing between last scroll height and new scroll hight 
        if l_ht==n_ht:
            break
        l_ht=n_ht
```

```python
search("fruits") #passing fruits into the fun
fruits=[]
#scraping images of fruits 
try:
    f=driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    for i in f:
        fruits.append(i.get_attribute("src"))
except:
    fruits.append("--")
```

```python
search("cars")
#scraping images of cars 

c=driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
for i in c:
    cars.append(i.get_attribute("src"))
```

```python
search("Machine Learning")
ML=[]

#scraping images of ML 

m=driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
for i in m:
    ML.append(i.get_attribute("src"))
```

```python
#data frame 
data=list(zip(fruits,cars,ML))
images=pd.DataFrame(data=data,columns=["Fruits","Cars","Machine Learning"])
images=images[0:100]
images
```

Q10 - Write a python program to scrape a data for all available Hostels from 
https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, 
distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms 
from price, facilities and property description.

```python
driver=webdriver.Chrome(r"C:\Users\Hp\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("https://www.hostelworld.com/")
```

```python
from time import sleep
s_bar=driver.find_element_by_xpath("//input[@class='location-text']")
s_bar.send_keys("London")

sleep(5)

z=driver.find_element_by_xpath("//div[@class='label'][1]")
z.click()
```

```python
s_button=driver.find_element_by_xpath("//button[@class='button primary large']")
s_button.click()
```

```python
page=1
name=[]
dist=[]
rating=[]
review=[]
p_price=[]
d_price=[]
fac=[]
desc=[]
url=[]

while page<3:
    
    #scraping distance from city centre
    d=driver.find_elements_by_xpath("//span[@class='description']")
    for i in d:
        dist.append(i.text)
        
    #Scraping privates from price
    try:
        pri=driver.find_elements_by_xpath("//a[@class='prices']/div[1]")[1:]
        for i in pri:
            p_price.append(i.text.replace("Privates From\n",""))
            
    except:
        p_price.append("--")
    
    
    #scraping dorm from price
    try:
        dor=driver.find_elements_by_xpath("//a[@class='prices']/div[2]")
        for i in dor:
            d_price.append(i.text.replace("Dorms From\n",""))
            
    except:
        d_price.append("--")
        
    
    #scraping url
    u=driver.find_elements_by_xpath("//h2[@class='title title-6']/a")
    for i in u:
        url.append(i.get_attribute("href"))
        
    page=page+1 #increment 
    
    #next icon
    try:
        n=driver.find_element_by_xpath("//div[@class='pagination-item pagination-last']")
        n.click()
    except:
        break
```

```python
for i in url:
    driver.get(i)
    
    #scraping hostel name 
    try:
        n=driver.find_element_by_xpath("//div[@class='title-2']")
        name.append(n.text)
    except:
        name.append("--")
    
    
    #scraping rating
    try:
        r=driver.find_elements_by_xpath("//div[@class='score orange big']")[1]
        rating.append(r.text)
    except:
        rating.append("--")
    
    
    #no.of reviews
    try:
        nr=driver.find_elements_by_xpath("//div[@class='reviews']")[1]
        review.append(nr.text)
    except:
        review.append("--")
        
        
    #scraping facilities 
    try:
        f=driver.find_element_by_xpath("//ul[@class='facility-sections']")
        fac.append(f.text)
    except:
        fac.append("--")
        
        
    #scraping property description 
    try:
        pd=driver.find_element_by_xpath("//div[@class='content collapse-content']")
        desc.append(pd.text)
    except:
        desc.append("--")
        
```

```python
import pandas as pd
#data frame 
data=list(zip(name,
dist,
rating,
review,
p_price,
d_price,
fac,
desc,
url,
))
hostel=pd.DataFrame(data=data,columns=["Hostel name","Distance from city centre","Rating","Total Reviews","Private from price","Droms from price","Facilities","Property description","URL"])
hostel
```
