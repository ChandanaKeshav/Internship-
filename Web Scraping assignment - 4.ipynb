Q10- Details of Datasets from UCI machine learning repositories.
 Url = https://archive.ics.uci.edu/
 You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver=webdriver.Chrome(r"C:\Users\Hp\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("https://archive.ics.uci.edu/ml/index.php")
```

```python
url=[]
u=driver.find_elements_by_xpath("//span[@class='normal']//a[1]")[6:30]
for i in u:
    url.append(i.get_attribute("href"))
    
url
```

```python
name=[]
datatype=[]
task=[]
attritype=[]
inst=[]
attri_no=[]
year=[]


for i in url:
    driver.get(i)
    
    #scraping the name of the dataset
    try:
        n=driver.find_element_by_xpath("//span[@class='heading']//b")
        name.append(n.text)
    except:
        name.append("--")
    
    
    
    #scraping data type
    try:
        dt=driver.find_elements_by_xpath("//tbody//td[2]//p")[0]
        datatype.append(dt.text)
    except:
        datatype.append("--")
    
    
    
    #scraping task
    try:
        t=driver.find_elements_by_xpath("//tbody//td[2]//p")[2]
        task.append(t.text)
    except:
        task.append("--")
        
        
    
    #scraping attribute type
    try:
        at=driver.find_elements_by_xpath("//tbody//td[2]//p")[1]
        attritype.append(at.text)
    except:
        attritype.append("--")
        
        
    
    #scraping number of instances 
    try:
        ni=driver.find_elements_by_xpath("//tbody//td[4]//p")[0]
        inst.append(ni.text)
    except:
        inst.append("--")
        
        
        
    #scraping number of attributes 
    try:
        nat=driver.find_elements_by_xpath("//tbody//td[4]//p")[1]
        attri_no.append(nat.text)
    except:
        attri_no.append("--")
        
        
    #scraping the year
    try:
        y=driver.find_elements_by_xpath("//tbody//td[6]//p")[1]
        year.append(y.text)
    except:
        year.append("--")
        
```

```python
#datatype
data=list(zip(name,datatype,task,attritype,inst,attri_no,year))
rep=pd.DataFrame(data=data,columns=["Data set","Data type","Task","Attribute type","No.of instance","no.of attribute","Year"])
rep
```
